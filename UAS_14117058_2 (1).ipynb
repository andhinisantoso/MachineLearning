```python
DATA = pd.get_dummies(AAAI, columns=['status_type', 'status_published'])
DATA
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(DATA)
```

```python
scaler.transform(DATA)
```

```python
data_normalized = pd.DataFrame(scaler.transform(DATA), columns=DATA.columns)
data_normalized.head()
```

```python
del data_normalized['Column1']
del data_normalized['Column2']
del data_normalized['Column3']
del data_normalized['Column4']
data_normalized.head()
```

```python
del data_normalized['Column1']
```

```python
data_normalized.head()
```

```python
from sklearn.cluster import KMeans

kmean = KMeans(n_clusters = 3)
```

```python
kmeans.fit(data_normalized)
```

```python
kmeans.predict(data_normalized)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(data_normalized, kmeans.predict(data_normalized))
```

```python
data = pd.get_dummies(data_initial, columns= ["type", "is_holiday"])
data
```

```python
data = data.drop('date', axis=1)
```

```python
predictor = ['store', 'department', 'temperature_c', 'unemployment', 'type_A', 'type_B', 'is_holiday_False', 'is_holiday_True']
target = ['weekly_sales']

X = data[predictor]
y = data[target]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
lr_model.predict(X_test)
```

```python
y_test
```

```python
y_pred = lr_model.predict(X_test)
y_pred
```

```python
print("MAE Score :", metrics.mean_absolute_error(y_test,y_pred))
print("MSE Score :",metrics.mean_squared_error(y_test,y_pred))
print("RMSE : ", metrics.mean_squared_error(y_test,y_pred, squared=False))
print("R2 Score :" , metrics.r2_score(y_test,y_pred))
```

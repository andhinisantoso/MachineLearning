```python
import pandas as pd

HCV = pd.read_csv("/content/drive/MyDrive/UAS/HCV data/hcvdat0.csv")
HCV.head()
```

```python
HCV.describe()
```

```python
HCV.info()
```

```python
HCV.isna().sum()
```

```python
mean_alb = round(HCV["ALB"].mean())
HCV["ALB"] = HCV["ALB"].fillna(mean_alb)

mean_alp = round(HCV["ALP"].mean())
HCV["ALP"] = HCV["ALP"].fillna(mean_alp)

mean_alt = round(HCV["ALT"].mean())
HCV["ALT"] = HCV["ALT"].fillna(mean_alt)

mean_chol = round(HCV["CHOL"].mean())
HCV["CHOL"] = HCV["CHOL"].fillna(mean_chol)

mean_prot = round(HCV["PROT"].mean())
HCV["PROT"] = HCV["PROT"].fillna(mean_prot)
```

```python
HCV.isna().sum()
```

```python
HCV = HCV.drop('Unnamed: 0', axis = 1)
HCV.head()
```

```python
predictor = ['Age', 'ALB', 'ALP', 'AST', 'BIL', 'CHE', 'CHOL', 'CREA', 'GGT', 'PROT']
target = ['Sex']

x = HCV[predictor]
y = HCV[target]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier

# definisikan model
model_knn = KNeighborsClassifier(n_neighbors=3)

# training
model_knn.fit(x_train, y_train)
```

```python
model_knn.predict(x_test)
```

```python
model_knn.score(x_test, y_test)
```

```python
from sklearn.svm import SVC 

model_svm = SVC(kernel='poly', degree=3) 

model_svm.fit(x_train, y_train) 
```

```python
model_svm.predict(x_test) 
```

```python
model_svm.score(x_test, y_test)
```
